```python
needed_countries= ['USA', 'JPN', 'DEU', 'CHE', 'GBR', 'FRA', 'ITA', 'ESP', 'CAN', 'NLD', 'IND', 'CHN', 'TUR', 'IDN', 'KOR', 'MEX', 'CHL', 'BRA', 'ISR', 'ZAF']
developed_countries = ['USA', 'JPN', 'DEU', 'CHE', 'GBR', 'FRA', 'ITA', 'ESP', 'CAN', 'NLD']
developing_countries = ['IND', 'CHN', 'TUR', 'IDN', 'KOR', 'MEX', 'CHL', 'BRA', 'ISR', 'ZAF']
```

```python
health_df = health_fin_df.copy()
```

```python
def country_status(value):
    if value in developed_countries:
        return "Developed"
    elif value in developing_countries:
        return "Developing"
 
health_df['dev_status'] = health_df['country_code'].map(country_status)
display(health_df.tail())
```

```python
#bringing dev_status column to front
dev = health_df['dev_status']
health_df.drop(labels=['dev_status'], axis=1,inplace = True)
health_df.insert(4, 'dev_status', dev)
```

```python
#dataframes for developed and developing countries for year 2018
developed_2018 = health_df.loc[(health_df['dev_status']=='Developed') & (health_df['year'] == 2018)]
developing_2018 = health_df.loc[(health_df['dev_status']=='Developing') & (health_df['year'] == 2018)]
```

```python
import plotly.graph_objects as go
import plotly.express as px
```

```python
health_df = health_df.replace(',','', regex=True)
```

```python
health_df = health_df.loc[:,~health_df.columns.duplicated()].copy()
```

```python
health_df = health_df.dropna(subset=['Gdppercapita', 'life_exp', 'che_gdp', 'population'])
```

```python
health_df.sort_values(by=['year'],inplace=True)
```

```python
#question 1
#health care spendings for all countries

fig = px.choropleth(health_df, locations="country_code",
                    color="che_gdp",
                    hover_name="country",
                    animation_frame="year",
                    title="Worldwide Health Expenditure",
                    labels={
                     "che_gdp": "Health Exp%",
                 },
                    #color_continuous_scale=px.colors.sequential.Plasma)
                    color_continuous_scale=px.colors.sequential.Blues)
fig.show()

#global financial crisis can be observed between 2006-2009
```

```python
filtered_country_df = health_df.loc[health_df['country_code'].isin(needed_countries)]
filtered_country_df.sort_values(by=['dev_status', 'country'],inplace=True)
```

```python
#question 2
#comparing private vs govt health expenditure for all countries

px.bar(
    data_frame = filtered_country_df,
    x = "country",
    y = ["pvtd_che","gghed_che"],
    opacity = 0.9,
    orientation = "v",
    barmode = 'group',
    title='Private Healthcare Expenditure',
    animation_frame="year", 
    animation_group="country",
    color_discrete_map={
        'pvtd_che': 'rgb(55, 83, 109)',
        'gghed_che': 'rgb(26, 118, 255)'
    },
    labels={
               "pvtd_che": "Private Expenditure",
               "gghed_che": "Government Expenditure"
                 }
)
```

```python
from statistics import mean
oops_mean = filtered_country_df.groupby(['dev_status', 'year'])['oops_che'].mean()
oops_mean_df = oops_mean.to_frame()
oops_mean_df = oops_mean_df.reset_index()
```

```python
#question 3
#aggredated comparison of out-of-pocket spending for developed vs developing countries

fig = px.line(oops_mean_df, x="year", y="oops_che", color='dev_status',labels={
                     "year": "Year",
                     "oops_che": "Out-of-pocket % of total health expenditure",
                     "dev_status": "Development Status"
                 },title="Out-of-pocket Expenditure")
#fig.data[0].line.color = 'rgb(55, 83, 109)'
#fig.data[1].line.color = 'rgb(26, 118, 255)'
fig.show()
```
